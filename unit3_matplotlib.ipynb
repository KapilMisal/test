```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[2,3,4,5]
y=[4,9,16,25]
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,10,0.1)
y=np.sin(x)
plt.figure(figsize=(3,3))

plt.plot(x,y)
plt.show()
```

```python
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(2,1,1)
plt.plot(x,y)

x=np.array([0,1,2,3])
y=np.array([10,12,14,16])
plt.subplot(2,1,2)
plt.plot(x,y)

plt.show()
```

```python
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(1,2,1)
plt.plot(x,y)

x=np.array([0,1,2,3])
y=np.array([10,12,14,16])
plt.subplot(1,2,2)
plt.plot(x,y)

plt.show()
```

```python
x=np.array([3,8,1,10])
plt.plot(x,marker='o')
plt.show()
```

```python
x=np.array([3,8,1,10])
plt.plot(x,marker='o',color='red',linestyle='dotted')
plt.show()
```

```python
x=np.array([3,8,1,10])
plt.plot(x,marker='o',color='red',linestyle='solid',ms=20,mec='black')
plt.show()
```

```python
x=np.arange(1,10,0.2)
y=np.sin(x)
plt.plot(x,y,marker='o',ms=5,mec='red',linestyle='dotted')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x=[5,2,9,4,7]
y=[10,5,8,4,2]
plt.bar(x,y,color='green')
plt.show()
```

```python
x=[5,2,9,4,7,3,8,10,1]
y=[10,5,8,4,2,1,6,3,9]
plt.scatter(x,y,marker='*',color='brown')
plt.show()
```

```python
cars=['Audi','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data=[23,17,35,29,12,14]
fig=plt.figure(figsize=(6,3))
plt.pie(data,labels=cars)
plt.show()
```

```python
x=[2,3,4]
y=[1,1.5,5]
plt.plot(x)
plt.plot(y)
plt.legend(["blue","green"],loc="upper left")
plt.show()
```

```python
x=np.random.normal(170,10,200)
plt.hist(x,color='red',bins=30)
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes(projection='3d')
z=np.linspace(0,1,100)
x=z*np.sin(25*z)
y=z*np.cos(25*z)
ax.plot3D(x,y,'green')
ax.set_title('3D Line plot')
plt.show()
```
